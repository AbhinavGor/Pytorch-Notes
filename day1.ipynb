### Normalize a tensor to range [0, 1]

```python
t1 = torch.rand((4, 4))
```

```python
t1 = t1 + 1000
t1 *= t1
# t1 = torch.exp(t1)
t1.view(2,8)
```

```python
t1_max = torch.max(t1)
t1_min = torch.min(t1)
```

```python
t1 = (t1 - t1_min)/(t1_max - t1_min)
```

```python
t1
```

### Compute column-wise mean and std of a 2D tensor.

```python
t2 = torch.rand((3, 6))
```

```python
t2
```

```python
for i in range(t2.shape[0]):
    print(t2[i:i+1])
```

```python
t2_col_means = [float(torch.mean(t2[:, i:i+1])) for i in range(t2.shape[1])]
```

```python
t2_col_means
```

```python
t2_col_means = t2.mean(dim=0)
t2_col_stds  = t2.std(dim=0)
```

```python
t2_col_means
```

```python
t2_col_stds
```
